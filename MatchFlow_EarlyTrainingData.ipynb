```python
import os
import re
import pandas as pd

data_dir = "."

# list of recap files
files = [f for f in os.listdir(data_dir) if f.endswith(".txt")]

input_texts = []
target_texts = []

def extract_recap_parts(text):
    # extracting basic elements
    match_score = re.search(r"Vassar College (Men|Women) \d+ v \d+ .*? (Men|Women)", text)
    tries = re.search(r"Tries: (.*?)(?:\n|Cons:|Conversions:)", text, re.IGNORECASE)
    conversions = re.search(r"Cons(?:ersions)?: (.*?)(?:\n|Cards:|Injuries:)", text, re.IGNORECASE)
    injuries = re.search(r"Injur(?:ed|ies): (.*?)(?:\n|JYA|Overall record|Sincerely)", text, re.IGNORECASE)
    notable_match = re.search(r"Poughkeepsie, NY - (.*?)Team:", text, re.DOTALL)

    # extracting fallback if anything is missing
    match_text = match_score.group(0).strip() if match_score else "Match data not found"
    tries_text = tries.group(1).strip().replace("\n", " ") if tries else "None"
    conv_text = conversions.group(1).strip().replace("\n", " ") if conversions else "None"
    injury_text = injuries.group(1).strip().replace("\n", " ") if injuries else "None"
    notable_text = notable_match.group(1).strip().replace("\n", " ") if notable_match else "No notable details extracted"

    # formatted input_text
    input_text = (
        f"Recap this game: Match: {match_text}. "
        f"Tries: {tries_text}. "
        f"Conversions: {conv_text}. "
        f"Cards: None. "  # May add card parsing in the future
        f"Injuries: {injury_text}. "
        f"Notable: {notable_text}."
    )

    # Using the full email minus the roster for target_text
    end_idx = text.find("Team:")
    target_text = text[:end_idx].split("\n")[-10:]
    target_text = " ".join([line.strip() for line in target_text if line.strip()])

    return input_text, target_text

# looping through all files
for file in files:
    with open(os.path.join(data_dir, file), "r") as f:
        text = f.read()
        input_txt, target_txt = extract_recap_parts(text)
        input_texts.append(input_txt)
        target_texts.append(target_txt)

# creating DataFrame
df = pd.DataFrame({
    "input_text": input_texts,
    "target_text": target_texts
})


df.to_csv("recaps.csv", index=False)
print("Saved to recaps.csv")
```
